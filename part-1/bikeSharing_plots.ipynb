```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dayData = pd.read_csv('bike+sharing+dataset/day.csv')
print(dayData.head())
```

```python
# Line Plot: Time Series of Bike Rentals Over Time
plt.figure(figsize=(10, 6))
dayData['dteday'] = pd.to_datetime(dayData['dteday'])
sns.lineplot(data=dayData, x='dteday', y='cnt')
plt.title('Bike Rentals Over Time')
plt.xlabel('date')
plt.ylabel('bike rentals')
```

```python
# Box Plot: Bike Rentals by Season
# Map season numbers to season names
season_labels = {1: 'Winter', 2: 'Spring', 3: 'Summer', 4: 'Fall'}
dayData['season_label'] = dayData['season'].map(season_labels)
sns.boxplot(data=dayData, x='season_label', y='cnt')
plt.title('Bike Rentals by Season')
plt.xlabel('Season')
plt.ylabel('Total Bike Rentals')
```

```python
print(dayData.head())
```

```python
# Bar Plot: Bike Rentals on Working vs Non-Working Days
status_labels = {0: 'non-working day', 1: 'working day'}
dayData['status_label'] = dayData['workingday'].map(status_labels)
sns.barplot(data=dayData, x='status_label', y='cnt')
plt.title('Bike Rentals on Working Days vs. Non-working Days')
plt.xlabel('Working Day')
plt.ylabel('Total Bike Rentals')
```

```python
# Scatter Plot: Temperature vs. Bike Rentals
sns.scatterplot(data = dayData, x='temp', y='cnt')
plt.title('Bike Rentals on Temperature')
plt.xlabel('Temperature')
plt.ylabel('Bike Rentals')
```
